## Obtener el lenguaje en que está escrita la página

```python
soup.find("html")
```

```python
lang = "spanish" if "es" in soup.find("html")["lang"] else "english"
```

```python
lang
```

---

```python
import re
```

```python
texto_original = re.sub(r"\[\d*\]", " ", texto_original)
texto_original = re.sub(r"\s+", " ", texto_original)
texto_limpio = re.sub(r"\W", " ", texto_original)
texto_limpio = re.sub(r"\s+", " ", texto_limpio)
texto_limpio = texto_limpio.lower()
```

```python
print(texto_original)
```

```python
print(texto_limpio)
```

```python
from nltk.corpus import stopwords

sw = stopwords.words(lang)
```

```python
len(sw)
```

```python
len(set(sw))
```

```python
import nltk
```

```python
len(nltk.sent_tokenize(texto_limpio))
```

```python
oraciones = nltk.sent_tokenize(texto_original)
palabras = nltk.word_tokenize(texto_limpio)
```

```python
len(palabras)
```

```python
palabras_limpias = [p for p in palabras if p not in sw]
```

```python
len(palabras_limpias)
```

```python
frecuencias = nltk.FreqDist(palabras_limpias)
```

```python
frecuencias.most_common(5)
```

```python
n_max = max(frecuencias.values())
N = 5 # Número de oraciones que llevará el resumen
```

```python
dicc_calif = defaultdict(int)
```

```python
for oracion in oraciones:
    oracion_tokenizada = nltk.word_tokenize(oracion.lower())
    if len(oracion_tokenizada) < 25:
        for p in oracion_tokenizada:
            if oracion not in dicc_calif.keys():
                dicc_calif[oracion] = frecuencias[p] / n_max
            else:
                dicc_calif[oracion] += frecuencias[p] / n_max
```

```python
dicc_calif = dict(dicc_calif) # Opcional
```

```python
dicc_calif.items()
```

```python
dicc_calif = dict(sorted(dicc_calif.items(), key=lambda item: item[1], reverse=True))
```

```python
resumen = "\n".join(list(dicc_calif.keys())[:N])
```

```python
print(resumen)
```
